# **プログラムの名称**

## **概要**

様々な占いができるプログラムです。一つ目「おみくじ」、二つ目「数秘術」、三つ目「5つの質問占い」、計三つの占いができます。
占い内容は今日の占いとラッキーアイテムや性格がわかるようなものにしました。

## **操作方法**

「おみくじ」ではCtrl + Enterを押すだけです。

「数秘術」では生年月日の入力と名前の入力をしてもらいます。このときの注意点は、生年月日は「xxxx-yy-zz」、名前は「ローマ字表記」で入力してください。

「5つの質問占い」では数字を入力するだけです。

## **制作者のコメント**

反省点：「数秘術」でこちら側が指定していない入力をした場合に「○○のようにしてもう一度入力してください」と表示できるようなプログラムを作成したかったです。

改善点：「数秘術」の名前入力をローマ字だけでなくひらがなやカタカナ、漢字でも占いができるようにすればよかったと思いました。

作成者の裏話：
「数秘術」についての話です。最初は漢字でもできるように合計画数を計算できるプログラムを作成しようとしてました。しかし、難しくなかなかうまく動作しなかったため、26文字で表記できるアルファベットを用いることにしました

感想：
私は占いが好きなので自分で占いを作成していて楽しかったです。私が作った3つの占い方法の中で「5つの質問占い」の作成が一番難しかったです。

```python
# おみくじ
%reset -f
import random

# 大吉のコメントリスト
a = [
    "今日は良いことが起こりそう。",
    "今日は新しい発見ができそう。",
    "今日は効率よく仕事ができそう。",
    "今日は素晴らしいアイデアが浮かびそう。"
]
b = [
    "水筒",
    "絆創膏",
    "新10000円札",
    "メモ帳"
]

#　中吉のコメントリスト
c = [
    "今日は楽しい一日になりそう。。",
    "今日は小さな幸せがありそう。",
    "今日は人を笑顔にできそう。",
]
d = [
    "ティッシュ",
    "鉛筆",
    "新5000円札",
    "リモコン"
]

# 小吉のコメントリスト
e = [
    "今日中にできることはやっておくほうが良いかも...。",
    "今日は昔の友達と会うと良いかも...。",
    "何も起こらない一日になりそう。",
]
f = [
    "ハサミ",
    "本",
    "新1000円札",
    "モバイルバッテリー"
]

# 末吉のコメントリスト
g = [
    "帰宅中は背後を警戒した方が良いかも...。",
    "今日は喧嘩中の友達と仲直りした方が良いかも...。",
    "今日は小さな怪我ができるかも...。",
]
h = [
    "CD",
    "墨汁",
    "お守り"
]

# 凶のコメントリスト
i = [
    "今日は日ごろの感謝を伝えると良いかも...。",
    "家から出るときは計画を立てて出かけた方が良いかも...。",
    "今日は災難な一日になりそう。",
]
j = [
    "日本地図",
    "お札"
]

# 1から100までの乱数を生成
random_number = random.randint(1, 100)
print(f"生成された乱数: {random_number}")

# 下一桁を取得
last_digit = random_number % 10

# 結果を判定
if last_digit == 7:
    print("大吉です。")
    comment = random.choice(a)
    print(comment)
    item = random.choice(b)
    print(f"ラッキーアイテム {item}")
elif last_digit in [6, 8]:
    print("中吉です。")
    comment = random.choice(c)
    print(comment)
    item = random.choice(d)
    print(f"ラッキーアイテム {item}")
elif last_digit in [4, 9]:
    print("小吉です。")
    comment = random.choice(e)
    print(comment)
    item = random.choice(f)
    print(f"ラッキーアイテム {item}")
elif last_digit in [2,3]:
    print("末吉です。")
    comment = random.choice(g)
    print(comment)
    item = random.choice(h)
    print(f"ラッキーアイテム {item}")
else:
    print("凶です。")
    comment = random.choice(i)
    print(comment)
    item = random.choice(j)
    print(f"ラッキーアイテム {item}")
```

```python
# 数秘術
import re

def calculate_life_path_number(birth_date):
    # 生年月日を数字に分解
    digits = list(map(int, birth_date.replace('-', '')))
    total = sum(digits)
    while total > 9:
        total = sum(map(int, str(total)))
    return total

def calculate_expression_number(name):
    # 各文字に対応する数字を定義
    letter_to_number = {
        'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5,
        'F': 6, 'G': 7, 'H': 8, 'I': 9, 'J': 1,
        'K': 2, 'L': 3, 'M': 4, 'N': 5, 'O': 6,
        'P': 7, 'Q': 8, 'R': 9, 'S': 1, 'T': 2,
        'U': 3, 'V': 4, 'W': 5, 'X': 6, 'Y': 7,
        'Z': 8
    }
    # 名前を大文字に変換
    total = sum(letter_to_number[char] for char in name.upper() if char in letter_to_number)
    while total > 9:
        total = sum(map(int, str(total)))
    return total

def get_number_meaning(number):
    meanings = {
        1: "リーダーシップ、独立、自立心",
        2: "協調、パートナーシップ、調和",
        3: "創造性、表現力、社交性",
        4: "安定性、実務能力、勤勉",
        5: "自由、冒険、変化",
        6: "家庭、愛、責任感",
        7: "内省、知恵、精神性",
        8: "権力、成功、物質的な豊かさ",
        9: "人道主義、自己犠牲、慈愛"
    }
    return meanings.get(number, "不明な番号")

# 生年月日の入力
birth_date = "2007-07-07"  #@param {type:"string"}
while not re.match(r'^\d{4}-\d{2}-\d{2}$', birth_date):
    birth_date = input("フォーマットが正しくありません。もう一度生年月日を入力してください (YYYY-MM-DD): ")

life_path_number = calculate_life_path_number(birth_date)
print(f"ライフパスナンバー: {life_path_number} - {get_number_meaning(life_path_number)}")

# 名前の入力
name = "OshimaHideaki"  #@param {type:"string"}
while not all(char.isalpha() or char.isspace() for char in name):
    name = input("ローマ字で記入してください。もう一度名前を入力してください: ")

expression_number = calculate_expression_number(name)
print(f"エクスプレッションナンバー: {expression_number} - {get_number_meaning(expression_number)}")
```

```python
%reset -f
# 質問占い

def ask_question(question, choices):
    print(question)
    for i, choice in enumerate(choices):
        print(f"{i + 1}: {choice['text']} (得点: {choice['score']})")

    while True:
        answer = input("あなたの選択肢を入力してください (1-{}): ".format(len(choices)))
        if answer.isdigit() and 1 <= int(answer) <= len(choices):
            return choices[int(answer) - 1]['score']  # 選択肢の得点を返す
        else:
            print("無効な選択肢です。もう一度入力してください。")

def calculate_result(total_score):
    if total_score <= 5:
        return "あなたは非常に内向的で、静かな時間を好むタイプです。"
    elif total_score <= 10:
        return "あなたはバランスの取れた性格で、社交的でもあり、内向的でもあります。"
    elif total_score <= 14:
        return "あなたは社交的で、友達と過ごすことが好きなタイプです。"
    else:
        return "あなたは非常に外向的で、常に新しい経験を求める冒険家です。"

def main():
    questions = [
        "1. あなたはどちらの活動が好きですか？",
        "2. 友達と過ごすとき、あなたはどう感じますか？",
        "3. 新しい人と出会うのは好きですか？",
        "4. 休日の過ごし方は？",
        "5. あなたの理想の仕事は？"
    ]

    choices = [
        [
            {"text": "一人の時間を楽しむ", "score": 1},
            {"text": "家族でお出かけ", "score": 2},
            {"text": "友達と遊びに行く", "score": 3}
        ],
        [
            {"text": "リラックスする", "score": 4},
            {"text": "楽しい", "score": 3},
            {"text": "緊張する", "score": 2},
            {"text": "話したくない", "score": 1}
        ],
        [
            {"text": "はい、好きです！", "score": 3},
            {"text": "あまり好きではない", "score": 1},
            {"text": "どちらとも言えない", "score": 2}
        ],
        [
            {"text": "家でのんびり", "score": 1},
            {"text": "友達と出かける", "score": 3},
            {"text": "趣味に没頭する", "score": 2}
        ],
        [
            {"text": "研究職", "score": 1},
            {"text": "営業職", "score": 3},
            {"text": "クリエイティブな仕事", "score": 2}
        ]
    ]

    total_score = 0

    for i in range(len(questions)):
        score = ask_question(questions[i], choices[i])
        total_score += score

    result = calculate_result(total_score)
    print("\n占い結果:")
    print(result)

if __name__ == "__main__":
    main()
```
